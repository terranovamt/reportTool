```python
uty.write_log("Write conclusion")

# Lists to store tests that pass and fail
pass_tests, fail_tests = [], []
pass_tests_results_df, fail_tests_results_df = [], []

# Filter columns in statistics
statistics = statistics[[
    "TestName",
    "Tested Part",
    "Low Limit",
    "Min",
    "Mean",
    "Max",
    "High Limit",
    "Std",
    "Unit",
    "Cp",
    "CpK",
    "Kurtosis",
    "Fail Duts",
    "Yield",
]]

# Check CpK value and Kurtosis in statistics
for _, row in statistics.iterrows():
    if ":Trimmed" in row["TestName"]:
        if row["Kurtosis"] != "-" and float(row["Kurtosis"]) > -1:
            fail_tests.append(row)
        else:
            pass_tests.append(row)
    else:
        if row["CpK"] != "-" and float(row["CpK"]) < 1.6:
            fail_tests.append(row)
        else:
            pass_tests.append(row)

# Check tests in results_df
for _, row in results_df.iterrows():
    percent_pass_value = float(row["Yield"].strip('%'))
    # print(percent_pass_value)
    (fail_tests_results_df if percent_pass_value < 75 else pass_tests_results_df).append(row)

# Determine the overall conclusion
all_tests_within_spec = not fail_tests and not fail_tests_results_df
conclusion_html = "<h1 id='conclusions'>Conclusions</h1>"
conclusion_html += "<h3>{}</h3>".format(
    "All tests are within specifications."
    if all_tests_within_spec
    else f"{len(fail_tests) + len(fail_tests_results_df)} out of {statistics.shape[0] + results_df.shape[0]} tests are not within specifications."
)

# Tests that fail
if fail_tests or fail_tests_results_df:
    conclusion_html += "<h2>Fail Tests:</h2>"
    if fail_tests:
        conclusion_html += "<h3>Parametric Tests statistics</h3>"
        conclusion_html += (
            statistics[
                ((statistics["CpK"].apply(lambda x: x != "-" and float(x) < 1.6)) & (~statistics["TestName"].str.contains(":Trimmed"))) |
                ((statistics["TestName"].str.contains(":Trimmed")) & (statistics["Kurtosis"].apply(lambda x: x != "-" and float(x) > -1)))
            ]
            .set_index("TestName")
            .style.map(uty.color_cpk, subset=["CpK"])
            .format(precision=3)
            .to_html(index=False, escape=False, index_names=False, columns=[
                "TestName", "Tested Part", "Low Limit", "High Limit", "Min", "Max", 
                "Mean", "Std", "Unit", "Cp", "CpK", "Kurtosis", "Fail Duts", "Yield"
            ])
        )
    if fail_tests_results_df:
        conclusion_html += "<h3>Functional Tests results</h3>"
        conclusion_html += results_df[results_df["Yield"].apply(lambda x: float(x.strip('%')) < 75)].to_html(index=False)

# Tests that pass
conclusion_html += "<h2>Pass Tests:</h2>"
if pass_tests:
    conclusion_html += "<h3>Parametric Tests statistics</h3>"
    conclusion_html += statistics[
        ((statistics["CpK"].apply(lambda x: x == "-" or float(x) >= 1.6)) & (~statistics["TestName"].str.contains(":Trimmed"))) |
        ((statistics["TestName"].str.contains(":Trimmed")) & (statistics["Kurtosis"].apply(lambda x: x == "-" or float(x) <= -1)))
    ].to_html(index=False, escape=False)
if pass_tests_results_df:
    conclusion_html += "<h3>Functional Tests results</h3>"
    conclusion_html += results_df[results_df["Yield"].apply(lambda x: float(x.strip('%')) >= 75)].to_html(index=False)

# Display conclusions
display(HTML(conclusion_html))
```
