```python
results = []
total_parts = 0
total_failed_parts = 0

discontinuities = [0] + [i + 1 for i in range(len(prr) - 1) if prr['SITE_NUM'].iloc[i + 1] != prr['SITE_NUM'].iloc[i] + 1] + [len(prr)]

for i in range(len(discontinuities) - 1):
    subset = prr.iloc[discontinuities[i]:discontinuities[i + 1]]
    
    total_parts += len(subset)
    total_failed_parts += len(subset[subset['HARD_BIN'] > 2])
    
    yield_percentage = (total_parts - total_failed_parts) / total_parts
    good_parts = total_parts - total_failed_parts
    results.append((total_parts, yield_percentage, total_failed_parts, good_parts))

results_df = pd.DataFrame(results, columns=['DUTs', 'Yield', 'Fail', 'Good'])

total_good_parts = results_df['Good'].max()
total_failed_parts = results_df['Fail'].max()
```
